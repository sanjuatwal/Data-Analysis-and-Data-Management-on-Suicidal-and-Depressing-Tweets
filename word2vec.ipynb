```python
from gensim.models import KeyedVectors
```

```python
model = KeyedVectors.load_word2vec_format("GoogleNews-vectors-negative300.bin", binary=True)
```

```python
causes = {'previous_suicide_attempts' : ['attempted', 'previously', 'tried', 'attempt', 'earlier', 'myself'],
'substance_misuse' : ['drugs', '420', 'weed', 'cocaine', 'pills', 'opiod', 'painkillers', 'alcoholic', 'heroine', 'meth'],
'access_to_lethal_means' : ['guns', 'ammo', 'ammunition', 'firearms', 'knife', 'pills', 'bullet'],
'losses_and_other_events' : ['breakup', 'death', 'died', 'left', 'divorce', 'separation', 'separated', 'fail', 'failure', 'study', 'studies', 'broke', 'financial', 'poor', 'lawsuit', 'court', 'bully', 'bullying', 'bullied', 'job'],
'history_of_trauma_or_abuse' : ['rape', 'raped', 'harassed', 'harass', 'harassment', 'trauma', 'assaulted', 'groped', 'molested'],
'chronic_physical_illness' : ['chronic','concussion', 'injury', 'head injury','back pain', 'gallbladder','ulcer', 'disabled','epilepsy','spinal','liver','bone disorder','joint disorder','gastrointestinal','cardiovascular','STD','blind','Deaf''renal disease', 'chronic obstructive pulmonary disease','COPD','cancer','human immunodeficiency virus','HIV','acquired immunodeficiency syndrome','AIDS','quadriplegia','multiple sclerosis','chronic heart failure','chronic pain','heart failure','kidney failure','dysphagia'],
'mood_disorders' : ['depression', 'bipolar disorder', 'bipolar', 'disorder', 'PTSD','post traumatic stress disorder','autism','autistic','schizophrenia',"aspergers", 'bipolar','ADHD''ADD''attention deficit','OCD','obsessive compulsive','depressive','panic attack','pedophile','pedophilia', 'paraphilia','mental illness','mental illnesses','diagnosis','borderline personality','bordeline','diagnosed', 'GAD','dissociative identity','dissociative amnesia', 'multiple personalities','psychosis','hallucinations','delusions','intellectual disability','intellectualy disabled', 'learning disability', 'binge eating','anorexia','anorexic','bulimia','bulimic','body dysmorphia']}
```

```python
word2vec_data = {}
for key in causes.keys():
  word2vec_data[key] = {}
  for inner_key in causes[key]:
    word2vec_data[key][inner_key] = {}
```

```python
for key in word2vec_data.keys():
  for inner_key in word2vec_data[key].keys():
    try:
      word2vec_data[key][inner_key] = model.wv.similar_by_word(inner_key)
    except:
      word2vec_data[key][inner_key] = 'delete'
```

```python
for key in word2vec_data:
  for inner_key in list(word2vec_data[key]):
    if word2vec_data[key][inner_key] == 'delete':  
      del word2vec_data[key][inner_key]
```
